```python
import pandas as pd
```

```python
nyc_listings = pd.read_csv('data/nyc_listings/listings.csv')
```

```python
nyc_listings.columns
```

```python
X = nyc_listings[[
       'host_since', 'host_location',
       'host_response_time', 'host_response_rate', 'host_acceptance_rate',
       'host_is_superhost',
       'host_neighbourhood', 'host_listings_count',
       'host_total_listings_count',
       'host_has_profile_pic', 'host_identity_verified',
       'neighbourhood_cleansed', 'neighbourhood_group_cleansed',
       # 'latitude', 'longitude',
       'property_type', 'room_type', 'accommodates', 'bathrooms',
       'bathrooms_text', 'bedrooms', 'beds', 'amenities',
       'minimum_nights', 'maximum_nights',
       # 'minimum_minimum_nights', 'maximum_minimum_nights', 'minimum_maximum_nights',
       # 'maximum_maximum_nights', 'minimum_nights_avg_ntm', 'maximum_nights_avg_ntm',
       'has_availability',
       # 'availability_30', 'availability_60', 'availability_90', 'availability_365',
       'number_of_reviews',
       # 'number_of_reviews_ltm', 'number_of_reviews_l30d',
       # 'first_review', 'last_review',
       'review_scores_rating', 'review_scores_accuracy',
       'review_scores_cleanliness', 'review_scores_checkin',
       'review_scores_communication', 'review_scores_location',
       'review_scores_value', 'instant_bookable',
       'calculated_host_listings_count',
       # 'calculated_host_listings_count_entire_homes', 'calculated_host_listings_count_private_rooms', 'calculated_host_listings_count_shared_rooms',
       'reviews_per_month']]

y = nyc_listings['price']
```

```python
# Convert percentages to numbers
for feature in ['host_response_rate', 'host_acceptance_rate']:
    X[feature] = X[feature].str.replace('%', '').astype(float) / 100
```

```python
# Convert t/f to integers
for feature in ['host_has_profile_pic', 'host_is_superhost', 'host_identity_verified', 'has_availability', 'instant_bookable']:
    X[feature] = X[feature].map({'t': 1, 'f': 0})
```

```python
# Convert bathroom text to bathroom type
X['bathroom_type'] = X['bathrooms_text'].apply(lambda x: 'private' if 'private' in str(x).lower() else ('shared' if 'shared' in str(x).lower() else None))
X.drop(columns=['bathrooms_text'], inplace=True)
```

```python
# Convert the strings to lists
import json
X['amenities'] = X['amenities'].apply(json.loads)

# Flatten the list and count occurrences
all_amenities = [amenity for sublist in X['amenities'] for amenity in sublist]
amenity_counts = pd.Series(all_amenities).value_counts()

# Get the top 20 most common amenities
top_20_amenities = amenity_counts.nlargest(20).index.tolist()

# Create new columns for only the top 20 amenities
for amenity in top_20_amenities:
    X[amenity] = X['amenities'].apply(lambda x: 1 if amenity in x else 0)

X.drop(columns=['amenities'], inplace=True)
```

```python
# Convert dates to floats
for feature in ['host_since', 'first_review', 'last_review']:   
    if feature in X.columns:
        # Convert the 'date_column' to datetime
        X[feature] = pd.to_datetime(X[feature])

        # Define the reference date
        reference_date = pd.to_datetime('2007-01-01')

        # Calculate the number of days since the reference date
        X[feature] = (X[feature] - reference_date).dt.days
```

```python
X.shape
```

```python
# Identify string columns
string_columns = X.select_dtypes(include=['object']).columns

# Get dummy variables for all string columns
X_dummies = pd.get_dummies(X, columns=string_columns, drop_first=True)
```

```python
X_dummies.shape
```

```python
# Convert prices from strings to floats
y = y.str.replace('[$,]', '', regex=True).astype(float)
```

```python
# Calculate the correlation with the target variable
correlation = X_dummies.corrwith(y)

# Sort the correlations by absolute value and select the top 30
top_features = correlation.abs().nlargest(30).index

# Create a new DataFrame with the top correlated features
X_top = X_dummies[top_features]
```
